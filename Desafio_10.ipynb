### Desafio 10

Dada uma lista, divida-a em 3 partes iguais e reverta a ordem de cada lista.

**Exemplo:** 

Entrada: \
sampleList = [11, 45, 8, 23, 14, 12, 78, 45, 89]

Saída: \
Parte 1 [8, 45, 11] \
Parte 2 [12, 14, 23] \
Parte 3 [89, 45, 78] 

```python
# Seu código
sampleList = [11, 45, 8, 23, 14, 12, 78, 45, 89]
```

```python
parte_1 = sampleList[0:3][ : : -1]
parte_1
```

```python
parte_2 = sampleList[3:6][ : : -1]
parte_2
```

```python
parte_3 = sampleList[6:9][ : : -1]
parte_3
```
